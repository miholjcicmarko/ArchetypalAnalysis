```python
data_states = data["id"]
data_states = set(data_states)
data_states = list(data_states)
```

```python
data_var = list(data.columns[2:])
data_var = [data_var[2]]
data_var
```

```python
states = data_states
variables = data_var

df = pd.DataFrame()

for i in range(len(states)):
    state = data.loc[data['id'] == ""+states[i]]
    newdf = state.iloc[0:1]
    
    for j in range(len(variables)):
        variable_name = variables[j]
        for k in range(state.shape[0]):
            var_name = str(k)
            value = state.iloc[k][""+variable_name]
            newdf[var_name+variable_name] = value
    
    df = pd.concat([df, newdf], ignore_index=True)
    
df
```

```python
df_data = df.fillna(0)
```

```python
df_data = df_data.drop(columns = list(data.columns))
df_data
```

```python
for i in range(7):
    noc_val = i + 1
    XC, S, C, SSE, varexpl = PCHA(df_num, noc=noc_val, I = None, delta=0)
    
    S_df = pd.DataFrame(S)
    XC_df = pd.DataFrame(XC)
    
    S_df.to_csv('/Users/markomiholjcic/Documents/GitHub/ArchetypalAnalysis/data/CovidSMatrix'+ \
                str(noc_val)+'.csv', index = False, header = True)
    XC_df.to_csv('/Users/markomiholjcic/Documents/GitHub/ArchetypalAnalysis/data/CovidXCMatrix'+ \
                 str(noc_val)+'.csv', index = False, header = True)
```
