```python
def check_float(potential_float):
    try:
        float(potential_float)
        return True
    except ValueError:
        return False
```

```python
a = data["embeddings"][0].split()
a[8]
```

```python
df_arch = pd.DataFrame()

for i in range(int(len(data["embeddings"])/100)):
    a = data["embeddings"][i].split()
    newdf = []
    for j in range(10):
        index = str(j)
        if (j == 0 and check_float(a[j][1:]) == True):
            newdf.append(a[j][1:])
        elif (j == 9 and check_float(a[j][:-1]) == True):
            newdf.append(a[j][:-1])
        elif (check_float(a[j])):
            newdf.append(a[j])
    newdf = np.array(newdf)
    newdf = pd.DataFrame(newdf)
    df_arch = pd.concat([df_arch, newdf], axis = 1, ignore_index=True)
    
df_arch
```
